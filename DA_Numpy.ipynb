```python
## Creating Arrays
```

```python
list1 = [1,2,3]
print(list1)
```

```python
arr_list1 = np.array(list1)
print(arr_list1)
```

```python
print(type(arr_list1))
```

```python
print(arr_list1.ndim)
```

```python
list1 = [1,2,3]
list2 = [4,5,6]
arr_list1_2 = np.array([list1  ,list2])
print(arr_list1_2)
```

```python
print(arr_list1_2.ndim)
```

```python
list1 = [1,2,3]
list2 = [4,5,6]
list3 = [7,8,9]
arr_list1_3 = np.array([[list1  ,list2,list3]])
print(arr_list1_3)
```

```python
print(arr_list1_3.ndim)
```

```python
## numpy array attribute
```

```python
arr1 = np.array([[1,2,3] , [4,5,6]])
print(arr1)
```

```python
print("Shape" , arr1.shape)
print("Size" , arr1.size)
print("dType" , arr1.dtype)
print("n_dim" ,arr1.ndim)
```

```python
## Array Initialization method
```

```python
#zeros Array
zero_arr = np.zeros((2,3))
print(zero_arr )
```

```python
## one array
one_arr = np.ones((1,3))
print(one_arr)
```

```python
## full array
full_arr = np.full((3,2),8)
print(full_arr)
```
